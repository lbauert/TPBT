## Gathering the datasets
Now we are going to get all of our split datasets and make a large list that we can use lapply on

```python
DL <- list('aa' = AA, 'ia' = IA, 'sn' = SN, 'bc' = BC, 'bi'= BI)
```

```python
DL
```

## Making a function to turn all the datasets into long form 

```python
lengthen <- function(x){
    x <- as.data.frame(x)
    names(x)<-c('Sweet', 'PID', 'w1_BMI', 'w1_BMI_status', 'age', 'ethnicity', 'race', 'gender', 'T1', 'T2', 'T3')
    x_long<-melt(data = x,
              id.vars = c('PID', 'Sweet','w1_BMI', 'w1_BMI_status', 'age', 'ethnicity', 'race', 'gender'),
              measure.vars = c('T1', 'T2', 'T3'))
    names(x_long)<-c('PID', 'Sweet','w1_BMI', 'w1_BMI_status', 'age', 'ethnicity', 'race', 'gender', 'time','measure')
    return(x_long)
}
```

## Using lapply
lapply functions like a loop would in python. It requires a list input (thus why we created that giant list). Now we can use our function on each item in the list

```python
DL_long <- lapply(DL, lengthen)
```

```python
DL_long
```

## Model with growth curve modeling
example of how we are going to model the data before we create a function

```python
AA_long
```

### Fit an intercept first
What does the variable of interest look like with no additional independent variables 

```python
intercept<-gls(attitude~1, data=AA_long, method="ML", na.action=na.exclude)
```

### Fit a random intercept
Here we are saying that every individual has their own intercept. 

```python
randomIntercept<-lme(
    attitude~1, data=AA_long, random=~1|PID, method="ML", na.action=na.exclude, control=list(opt="optim"))
```

### Fit a random intercept and a independent variable

```python
timeRI<-update(randomIntercept, .~.+time_bev)
```

### Fit a random slope and a random intercept
Here we are acknowledging that each subject has a random intercept and the slope is random over time

```python
timeRS<-update(timeRI, random=~time_bev|PID)
```

### Examine the results
Is anything interesting?

```python
summary(timeRS)
anova(timeRS)
```

### Fit an additional independent variable
Is there a difference between groups?

```python
timeCov<-update(timeRS, .~.+Sweet)
summary(timeCov)
anova(timeCov)
```

```python
anova(intercept, randomIntercept, timeRI, timeRS, timeCov)
```

```python
#x = data in long form
fitter<-function(x){
    x$Sweet<-as.factor(x$Sweet)
    intercept<-gls(measure~1, data=x, method="ML", na.action=na.exclude)
    randomIntercept<-lme(measure~1, data=x, random=~1|PID, method="ML", na.action=na.exclude, control=list(opt="optim"))
    timeRI<-update(randomIntercept, .~.+time)
    timeRS<-update(timeRI, random=~time|PID)
    timeCov<-update(timeRS, .~.+Sweet)
    results <-list("basic"=summary(timeCov), "better?"= anova(intercept, randomIntercept, timeRI, timeRS, timeCov))
    return(results)
}
```

```python
lapply(DL_long, fitter)
```

### Creating a function to speed up graphing 

```python
plotter<-function(x){
    x$Sweet<-as.factor(x$Sweet)
    plot1 <- ggplot(x, aes(time, measure, colour = Sweet))
    plot2<-plot1 + stat_summary(
        fun.y = mean, geom = "point") + stat_summary(
        fun.y =mean, geom = "line", aes(group= Sweet)) + stat_summary(
        fun.data = mean_cl_boot, geom   = "errorbar", width = 0.2) + labs(
        x = "time", y = "measure" , colour = "Sweet")    
    return(plot2)
}
```

### Using lapply to use the new plotter function on the whole dataset (as a list)

```python
lapply(DL_long, plotter)
```
